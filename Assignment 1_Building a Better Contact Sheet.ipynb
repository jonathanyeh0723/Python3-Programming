```python
# Import nessary libraries for the assignment
from PIL import Image, ImageDraw, ImageFont
```

```python
# Grab the base image
image = Image.open("readonly/msi_recruitment.gif")
image = image.convert('RGB')

# Create an empty list
list_img = []

# To check the image, let's display it
display(image)
```

```python
# Build the images

# Channel 0
for intensity in (0.1, 0.5, 0.9):
        
    r, g, b = image.split()
    new_r = r.point(lambda x: x * intensity)
    new_img = Image.merge('RGB', (new_r, g, b))
        
    black_rect = Image.new('RGB', (new_img.width, 50), color = 0)
    draw = ImageDraw.Draw(black_rect)
    fnt = ImageFont.truetype("readonly/fanwood-webfont.ttf", 50)
    draw.text((10, 10), 'channel 0 intensity {}'.format(intensity), font = fnt, fill = new_img.getpixel((0, 100)))
        
    sheet = Image.new('RGB', (new_img.width, new_img.height + black_rect.height))
    sheet.paste(new_img, (0, 0))
    sheet.paste(black_rect, (0, new_img.height))
        
    list_img.append(sheet)

# Channel 1
for intensity in (0.1, 0.5, 0.9):
        
    r, g, b = image.split()
    new_g = g.point(lambda x: x * intensity)
    new_img = Image.merge('RGB', (r, new_g, b))
        
    black_rect = Image.new('RGB', (new_img.width, 50), color = 0)
    draw = ImageDraw.Draw(black_rect)
    fnt = ImageFont.truetype("readonly/fanwood-webfont.ttf", 50)
    draw.text((10, 10), 'channel 1 intensity {}'.format(intensity), font = fnt, fill = new_img.getpixel((0, 100)))
        
    sheet = Image.new('RGB', (new_img.width, new_img.height + black_rect.height))
    sheet.paste(new_img, (0, 0))
    sheet.paste(black_rect, (0, new_img.height))
        
    list_img.append(sheet)

# Channel 2
for intensity in (0.1, 0.5, 0.9):
        
    r, g, b = image.split()
    new_b = g.point(lambda x: x * intensity)
    new_img = Image.merge('RGB', (r, g, new_b))
        
    black_rect = Image.new('RGB', (new_img.width, 50), color = 0)
    draw = ImageDraw.Draw(black_rect)
    fnt = ImageFont.truetype("readonly/fanwood-webfont.ttf", 50)
    draw.text((10, 10), 'channel 2 intensity {}'.format(intensity), font = fnt, fill = new_img.getpixel((0, 100)))
        
    sheet = Image.new('RGB', (new_img.width, new_img.height + black_rect.height))
    sheet.paste(new_img, (0, 0))
    sheet.paste(black_rect, (0, new_img.height))
        
    list_img.append(sheet)
```

```python
# Build a new contact sheet, and paste the images we create into it.

x = 0
y = 0

width, height = list_img[0].size

contact_sheet = Image.new('RGB', (width * 3, height * 3))

for image in list_img:
    
    contact_sheet.paste(image, (x, y))
    
    if x + width == contact_sheet.width:
        x = 0
        y += height
    else:
        x += width

# Resize to half of its original size
contact_sheet = contact_sheet.resize((int(contact_sheet.width / 2), int(contact_sheet.height / 2)))

# Finally, display that
display(contact_sheet)
```
